```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("lead_720_405.jpg")
```

```python
imgC = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imgC)
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml') 
```

```python
faces = face_cascade.detectMultiScale(imgC)
```

```python
faces
```

```python
BLACK = (0,0,0)
WHITE = (255,255,255)
for (x,y,w,h) in faces:
    cv2.rectangle(imgC, (x,y), (x+w,y+h), WHITE, 4, cv2.LINE_AA, 15)
    cv2.rectangle(imgC, (x,y), (x+w,y+h), BLACK, 2, cv2.LINE_AA)
```
